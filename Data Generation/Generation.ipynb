# Data Generation

```python
# Import libraies used 
import random
from random import randint
```

###  **Generate data input files:**

```python
# Read in data:
f = open("words_alpha.txt", "r")
words = []
for line in f.readlines():
    word = line.strip()
    words.append(word)

len(words)
```

```python
# Generate 7 different dataset sizes by sampling the data

words100000 = random.sample(words, 100000)
words50000 = random.sample(words100000, 50000)
words20000 = random.sample(words50000, 20000)
words10000 = random.sample(words20000, 10000)
words5000 = random.sample(words10000, 5000)
words1000 = random.sample(words5000, 1000)
words500 = random.sample(words1000, 500)

file_names = ['words5000','words50000' ,'words10000' ,'words20000','words100000' ,'words500' ,'words1000' ]
files = [words5000,words50000 ,words10000 ,words20000,words100000 ,words500 ,words1000]
```

###  **Generate operation input files:**

```python
for i,file in enumerate(file_names):
    with open(str(file)+'.txt', 'w+') as f:
        for word in files[i]:
            f.write(str(word))
            f.write(' ')
            value = randint(1,len(files[i]))
            f.write(str(value))
            f.write('\n')
        
        f.close()
```

```python
# Generate 50 random add commands 
words5 = random.sample(words, 50)
with open('testadd.in', 'w+') as f:
        for word in words5:
            f.write('A')
            value = randint(1,100000)
            f.write(' ')
            f.write(str(word))
            f.write(' ')
            f.write(str(value))
            f.write('\n')
        
        f.close()
```

```python
# Generate 50 random search commands 
words1 = random.sample(words, 50)
with open('testsearch.in', 'w+') as f:
        for word in words1:
            f.write('S')
            f.write(' ')
            f.write(str(word))
            f.write('\n')
        
        f.close()
```

```python
# Generate 50 random delete commands 
words2 = random.sample(words, 50)
with open('testdel.in', 'w+') as f:
        for word in words2:
            f.write('D')
            f.write(' ')
            f.write(str(word))
            f.write('\n')
        
        f.close()
```

```python
# Generate 50 random Autocomplete commands 
words3 = random.sample(words, 50)
with open('testAC.in', 'w+') as f:
        for word in words3:
            f.write('AC')
            f.write(' ')
            val = round(len(word)-(len(word)/2))
            value = randint(1,val)
            f.write(str(word[0:value]))
            f.write('\n')
        
        f.close()
```

```python
# Generate 50 random commands 
words4 = random.sample(words, 100)

with open('testfull.in', 'w+') as f:
            command = ['S','D','A','AC']
            count=0
            for c in command:
                
                for i in range(0,25):
                    w=i+25*count
                    word = words4[w]
                    f.write(c)
                    f.write(' ')
                    if c == 'AC':
                        val = round(len(word)-(len(word)/2))
                        value = randint(1,val)
                        f.write(str(word[0:value]))
                    else:
                        f.write(word)
                    if c =='A':
                        f.write(' ')
                        f.write(str(randint(1,100000)))

                    f.write('\n')
                count+=1
        
            f.close()
```
